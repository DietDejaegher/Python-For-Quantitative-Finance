```python
#merge multiple stock in one df by column name
def merge_df_by_column_name(col_name, sdate, edate, *tick_args):
    mult_df = pd.DataFrame()
    #print(tick_args[0])
    for x in tick_args:
        #print(x)
        df = get_df_from_csv(x)
        #df['Date']=pd.to_datetime(df['Date'])
        mask = (df.index >= sdate) & (df.index <= edate)
        mult_df[x] = df.loc[mask][col_name]
    
    return mult_df
        
```

```python
#Markowits Portfolio Optimization
#Optimizing return while minimizing risk
#Use uncorrelated stocks

#Most promising stocks from each sector
port_list = ["GNRC", "DXCM", "AMD", "NFLX", "COST", "TGT", "AES", "MSCI", "NEM", "AMT", "HES"]
```

```python
returns = np.log(mult_df / mult_df.shift(1))
mean_returns = returns.mean() * 252
mean_returns
```

```python
#Generate random weights equal total to 1
weights = np.random.random(len(port_list))
weights /= np.sum(weights)
print('Weights: ', weights)
print('Total Weights: ', np.sum(weights))
```

```python
#Calculate portfolio return
np.sum(weights * mean_returns) * 252
```

```python
#Calculate volatility of random portfolio
np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
```

```python
#Analyze returns and risk of 10000 random combinations
p_ret = []
p_vol = []
p_SR = []
p_wt = []

n_combinations = 10000

print('Number of assets in portfolio: ', len(port_list))
print('Number of random combinations tried with these assets: ', n_combinations)

for x in range(n_combinations):
    p_weights = np.random.random(len(port_list))
    p_weights /= np.sum(p_weights)

    ret_1 = np.sum(p_weights * returns.mean()) * 252
    p_ret.append(ret_1)
    
    vol_1 = np.sqrt(np.dot(p_weights.T, np.dot(returns.cov() * 252, p_weights)))
    #print("Vol_1: ", vol_1)
    p_vol.append(vol_1)

    SR_1 = (ret_1 - risk_free_rate)/vol_1
    p_SR.append(SR_1)
    
    p_wt.append(p_weights)

p_ret = np.array(p_ret)
p_vol = np.array(p_vol)
p_SR = np.array(p_SR)
p_wt = np.array(p_wt)

display(p_ret, p_vol, p_SR, p_wt)
```

```python
#Plot Efficient Frontier
ports = pd.DataFrame({'Return':p_ret, 'Vol':p_vol})
display(ports)
ports.plot(x='Vol', y='Return', kind='scatter', figsize=(12,5), title='Efficient Frontier')
```

```python
#Sharpe Ratio
SR_idx = np.argmax(p_SR)
i = 0
port_weights = []
while i < len(port_list):
    print("Stock: %s: %2.2f" % (port_list[i], (p_wt[SR_idx][i] * 100)))
    port_weights.append(p_wt[SR_idx][i] * 100)
    i += 1

rounded_port_weights = [round(x) for x in port_weights]

print("\nVolatility: ", p_vol[SR_idx])
print("\nReturn: ", p_ret[SR_idx])
print("Optimal portfolio weights: ", port_weights)
print("Rounded portfolio weights: ", rounded_port_weights)
```

```python
#create function that calculates round number of shares for the optimal portfolio
def get_port_shares(one_perc_price, force_one, wts, prices):
    num_stocks = len(wts)
    shares = []
    cost_shares = []

    i = 0
    while i < num_stocks:
        max_price = one_perc_price * wts[i]
        num_shares = int(max_price / prices[i])

        if (force_one & (num_shares == 0)):
            num_shares = 1
        shares.append(num_shares)
        cost = num_shares * prices[i]
        cost_shares.append(cost)
        i+=1

    return shares, cost_shares
```

```python
#Portfolio weighting
def get_port_weighting(share_cost):
    stock_wts = []
    tot_val = sum(share_cost)
    print("Total Investment = ", tot_val)

    for x in share_cost:
        stock_wts.append(x / tot_val)

    return stock_wts
```
